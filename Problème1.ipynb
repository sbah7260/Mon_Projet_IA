```python
df = m.drop("final_result", axis=1)
```

```python
data_x = m.loc[:, ["code_module","code_presentation","gender","region","highest_education","imd_band","age_band","disability"]]
```

```python
data_x
```

```python
encoder1 = OneHotEncoder(categories='auto')
x = encoder1.fit_transform(data_x)
p = pd.DataFrame(x.toarray())
```

```python
df = pd.concat([p, dummy_m], axis=1)
df
```

```python
x = df.iloc[:, :-4].values
```

```python
X = np.array(df.iloc[:, :-4])
y = df[["Distinction"]]
# y = df[["Fail"]]
# y = df[["Pass"]]
# y = df[["Withdrawn"]]
y
```

```python
#Entrainemment des données train et test
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2, random_state =5 )
# print(Y_test)
print(Y_test)
```

```python
# SVM régularization des paramétres
C = 1.0
g = 3.0
models = (
    svm.SVC(kernel="linear", C=C),
    svm.LinearSVC(C=C, max_iter=10000),
    svm.SVC(kernel="rbf", gamma=g, C=C),
    svm.SVC(kernel="poly", degree=3, gamma="auto", C=C),
)
```

```python
print(models)
```

```python
for clf in models:
    clf.fit(X_train, Y_train)
    yhat= clf.predict(X_test)
    print(" -Implementation", clf)
    print("Valeur predite:")
    print(yhat[:5])
    print("Ce qui est attendu:\n",Y_test[:5])

    print("Les métriques", clf)
    mse = mean_squared_error(Y_test, yhat)
    score = clf.score(X_test, Y_test)
```
